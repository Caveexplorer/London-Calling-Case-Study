**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
properties_T = properties.transpose()
```

```python
properties_T.head()
```

```python
properties_T.shape
```

```python
properties_T.index
#now we can see that the new indices of our transposed table are Borough names
```

```python
#reset the index to numbers, rather than the borough names
properties_T = properties_T.reset_index()
```

```python
properties_T.head()
```

```python
#Our columns are numbers, but we'd like them to be properties of the data. The properties happen to be in the first row of
#the data set.
new_header = properties_T.iloc[0]
properties_T = properties_T[1:]
properties_T.columns = new_header
```

```python
#Now the properties are the column names, except for two exceptions dealt with in the next cell.
properties_T.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#change the first to columns to London Boroughs and ID, respectively.
properties_T.rename(columns={'Unnamed: 0': 'London Boroughs', pd.NaT: 'ID'}, inplace = True)
```

```python
properties_T.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#"Melt" the long list of month columns into their own singular column
clean_properties = pd.melt(properties_T, id_vars = ['London Boroughs', 'ID'])
```

```python
clean_properties.head()
```

```python
clean_properties.shape
```

```python
#That new column for the dates needs a new name, 'Month.' We'll change the value column to be called
#'Average Price', since that's more descriptive.
clean_properties.rename(columns= {0: 'Month', 'value': 'Average Price'}, inplace = True)
```

```python
clean_properties.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
#Average price column is mixed data types. We want them to be floats so they can be mathematically operated on
clean_properties.dtypes
```

```python
#Convert all the entries in the Average price column to a numeric data type
average_price_column = pd.to_numeric(clean_properties['Average Price'], errors='coerce')
clean_properties['Average Price'] = average_price_column
clean_properties['Average Price'].fillna(0, inplace=True)
clean_properties['Average Price'] = pd.to_numeric(clean_properties['Average Price'])
```

```python
clean_properties.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
#Many of these are not London Boroughs, which we shoudn't concern ourselves with
boroughs = clean_properties['London Boroughs'].drop_duplicates()
boroughs
```

```python
#All entries after the last borough, Westminster, and the City of London entry, should be excluded.
list_to_drop = boroughs.tolist()
list_to_drop = list_to_drop[33:]
list_to_drop.append('City of London')
list_to_drop
```

```python
#These are the 32 boroughs we are concerned with
correct_boroughs = clean_properties[~clean_properties['London Boroughs'].isin(list_to_drop)]
correct_boroughs['London Boroughs'].drop_duplicates()
```

```python
#Index needed resetting
correct_boroughs.reset_index(drop=True)
```

```python
correct_boroughs.shape
```

```python
#All columns contain valid data types
correct_boroughs.isna().sum()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
correct_boroughs[correct_boroughs['London Boroughs'] == 'Barnet'].plot.line(x = 'Month', y = 'Average Price', 
                                                                            xlabel = 'Time (months)', ylabel = 'Home Price', 
                                                                            title = "Average Price over Time for the Barnet Borough")
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:

```python
#Having an average home price for each month for each borough in our two decade span borough was a little excessive. 
#Let's make a new column of just the years and add it to our dataframe.
years = pd.DatetimeIndex(correct_boroughs['Month']).year
list(years)
correct_boroughs['Year'] = years
correct_boroughs.tail()
```

```python
#Now that we have a column for just the year, it's time to make a new dataframe without the months column, but with the average 
#price per year for each borough, averaged from the average price data we had per month. A random sample of ten, just to 
#check.
price_per_year = correct_boroughs.groupby(by=['London Boroughs', 'Year']).mean()
price_per_year.sample(10)
```

```python
#Recover our lost index
price_per_year = price_per_year.reset_index()
price_per_year.head()
```

```python
#The last time we used .shape on correct_boroughs we had over ten thousand rows. Now, after condensing average price per month
#into average price per year, we're down to 864 rows.
price_per_year.shape
```

```python
price_per_year[price_per_year['London Boroughs'] == 'Barnet'].plot.line(x = 'Year', y = 'Average Price', 
                                                                            xlabel = 'Time (months)', ylabel = 'Home Price', 
                                                                            title = "Average Price over Time for the Barnet Borough")
plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#computes the ratio of average price per home in 1998 over 2021 for whatever borough it is called on
def create_price_ratio(d):
    y1998 = float(d['Average Price'][d['Year']==1998])
    y2021 = float(d['Average Price'][d['Year']==2021])
    ratio = [y1998/y2021]
    return ratio
```

```python
#testing on a borough
create_price_ratio(price_per_year[price_per_year['London Boroughs']=='Barking & Dagenham'])
```

```python
#fills a dictionary where the keys are boroughs and the values are price ratios computed from the create_price_ratio function
final = {}
for b in price_per_year['London Boroughs'].unique():
    borough = price_per_year[price_per_year['London Boroughs'] == b]
    final[b] = create_price_ratio(borough)
print(final) 
```

```python
#print the key (borough) and value (average price) associated with the minimum value of the final dictionary
print(min(final, key=final.get), final[min(final, key=final.get)])
```

```python
#The top ten most expensive boroughs in 2021
df = price_per_year[price_per_year['Year'] == 2021]
df.sort_values('Average Price', ascending = False).head(10)
```
